```python
data=[[2,0,81], [4,4,93],[6,2,91],[8,3,97]]
```

```python
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('score')
ax.scatter(x1, x2, y)
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)
```

```python
a1 = 0
a2 = 0
b = 0
lr = 0.05#learn rate
epochs=2001
```

```python
#반복하면서 경사 하강법 시작
for i in range(epochs):#epochs수만큼 반복 2000번pochs수만큼 반복 2000번
    y_pred = a1*x1_data+a2*x2_data+b# 예측값을 구한다.
    error = y_data - y_pred #정답값과 예측값사이의 차이 에러를 구한다.
    a1_diff=-(1/len(x1_data))*sum(x1_data*(error))
    a2_diff=-(1/len(x2_data))*sum(x2_data*(error))
    b_diff=-(1/len(x1_data))*sum(error)
    a1=a1-lr*a1_diff #학습률을 곱해 기존의 a1값 업데이트
    a2=a2-lr*a2_diff #학습률을 곱해 기존의 a2값 업데이트
    b=b-lr*b_diff #학습률을 곱해 기존의 b값 업데이트
    
    if i % 100 ==0:
        print('epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f' %(i, a1, a2, b))
```

```python
#앞서 구한 기울기와 절편을 이용해 그래프를 다시 그리기
y_pred = a1*x1_data+a2*x2_data+b
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('score')
ax.scatter(x1, x2, y)
ax.plot_surface([min(x1_data), max(x1_data)],[min(x2_data), max(x2_data)],[min(y_pred), max(y_pred)])#선 그래프
plt.show()
```
